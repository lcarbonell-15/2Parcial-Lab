## 1. SYMPY - INTEGRALES

### 1.1 Configuración básica

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Crear variable simbólica
x = sp.symbols('x')
```

### 1.2 Integral Indefinida

**Fórmula:** $\int f(x)\,dx = F(x) + C$

Derivar e integrar son operaciones inversas.

```python
# Ejemplo 1: integral de 3x^2
f = 3*x**2
F = sp.integrate(f, x)
print(f"\u222b {f} dx = {F} + C")
```

```python
# Ejemplo 2: integral de e^x
f = sp.exp(x)
F = sp.integrate(f, x)
print(f"\u222b {f} dx = {F} + C")
```

```python
# Ejemplo 3: integral con condición inicial
# F(0) = 5, donde IMg = 2x
f = 2*x
F_general = sp.integrate(f, x)
print(f"Integral general: {F_general} + C")

# Resolver para C usando condición F(0) = 5
C = sp.symbols('C')
sol = sp.solve(sp.Eq(F_general.subs(x, 0) + C, 5), C)
print(f"C = {sol[0]}")
print(f"Resultado: {F_general} + {sol[0]} = {F_general + sol[0]}")
```

### 1.3 Verificar con derivada

```python
# Verificar que la integral es correcta derivando
F = x**3
f_verificada = sp.diff(F, x)
print(f"d/dx({F}) = {f_verificada}")  # Debe dar 3x^2
```

### 1.4 Integral Definida

**Regla de Barrow:** $\int_a^b f(x)\,dx = F(b) - F(a)$

```python
# Integral definida: \int_0^3 x^2 dx = 9
f = x**2
resultado = sp.integrate(f, (x, 0, 3))
print(f"\u222b_0^3 x^2 dx = {resultado}")
```

### 1.5 Aproximación por Riemann

**Fórmula:** $\sum_{i=1}^{n} f(x_i) \cdot \Delta x$

```python
# Aproximación Riemann de \int_0^3 x^2 dx con n=5
n = 5
a, b = 0, 3
dx = (b - a) / n
xi = np.linspace(a, b, n + 1)[1:]  # puntos derecho
altura = xi**2
aprox = np.sum(altura * dx)
print(f"Aproximación Riemann (n={n}): {aprox:.2f}")
print(f"Valor exacto: 9")
```

## 2. APLICACIONES ECONÓMICAS CON INTEGRALES

### 2.1 Costo Marginal → Costo Total

$C(q) = \int CMg(q)\,dq + F$

```python
# Ejemplo: CMg = 20 + 2q, F = 100
q = sp.symbols('q')
CMg = 20 + 2*q
F_fijo = 100

C = sp.integrate(CMg, q) + F_fijo
print(f"C(q) = {C}")

# Costo medio
C_med = C / q
print(f"C_med(q) = {C_med}")

# Mínimo del costo medio: dC_med/dq = 0
dCmed = sp.diff(C_med, q)
q_min = sp.solve(dCmed, q)
print(f"q mínimo costo medio: {q_min}")
print(f"C_med mínimo: {C_med.subs(q, q_min[0])}")
```

### 2.2 Ingreso Marginal → Ingreso Total

$I(q) = \int IMg(q)\,dq$

No se agrega constante porque $I(0) = 0$.

```python
# Ejemplo: IMg = 200 - 4q
IMg = 200 - 4*q
I = sp.integrate(IMg, q)
print(f"I(q) = {I}")

# Máximo ingreso: IMg = 0
q_max = sp.solve(IMg, q)
print(f"q* ingreso máximo: {q_max[0]}")
print(f"I_max = {I.subs(q, q_max[0])}")
```

### 2.3 Beneficio y Punto Óptimo

$B(q) = I(q) - C(q)$

Óptimo: $BMg = IMg - CMg = 0$

```python
# Ejemplo: IMg = 250 - 10q, CMg = 50 + 2q, CF = 0
IMg_ej = 250 - 10*q
CMg_ej = 50 + 2*q

I_ej = sp.integrate(IMg_ej, q)
C_ej = sp.integrate(CMg_ej, q)
B_ej = I_ej - C_ej

print(f"I(q) = {I_ej}")
print(f"C(q) = {C_ej}")
print(f"B(q) = {B_ej}")

# Punto óptimo: BMg = 0
BMg = sp.diff(B_ej, q)
q_opt = sp.solve(BMg, q)
print(f"q* = {q_opt[0]}")
print(f"B* = {B_ej.subs(q, q_opt[0])}")
```

### 2.4 Excedente del Consumidor (EC)

$EC = \int_0^{q^*} [P_d(q) - p^*]\,dq$

```python
# Demanda: Qd = 120 - 3p → Pd = (120 - q)/3
# Oferta: Qs = -20 + 2p → Ps = (q + 20)/2
q_ec = sp.symbols('q_ec')
Pd = (120 - q_ec) / 3
Ps = (q_ec + 20) / 2

# Equilibrio: Pd = Ps
q_star = sp.solve(sp.Eq(Pd, Ps), q_ec)
p_star = Pd.subs(q_ec, q_star[0])
print(f"q* = {q_star[0]}, p* = {p_star}")

# EC
EC = sp.integrate(Pd - p_star, (q_ec, 0, q_star[0]))
print(f"EC = {EC}")

# EP
EP = sp.integrate(p_star - Ps, (q_ec, 0, q_star[0]))
print(f"EP = {EP}")
```

### 2.5 Beneficio Acumulado (Integral definida de BMg)

```python
# Beneficio acumulado entre q=2 y q=5
# IMg = 20-2x, CMg = 4+(x-4)^2, CF = 10
x_b = sp.symbols('x_b')
IMg_b = 20 - 2*x_b
CMg_b = 4 + (x_b - 4)**2

I_b = sp.integrate(IMg_b, x_b)
C_b = sp.integrate(CMg_b, x_b) + 10
B_b = I_b - C_b

B_acum = sp.integrate(IMg_b - CMg_b, (x_b, 2, 5))
print(f"Beneficio acumulado [2,5] = {B_acum}")
```

### 2.6 Utilidad Marginal

$U(q) = \int UMg(q)\,dq$

Máxima utilidad cuando $UMg = 0$.

```python
# UMg = 50 - 2q
UMg = 50 - 2*q
U = sp.integrate(UMg, q)
print(f"U(q) = {U}")

q_max_u = sp.solve(UMg, q)
print(f"Utilidad máxima en q = {q_max_u[0]}, U = {U.subs(q, q_max_u[0])}")
```

## 3. VALOR DEL DINERO EN EL TIEMPO

### 3.1 Equivalencia de Tasas

$i(t_2) = (1 + i(t_1))^{t_2/t_1} - 1$

```python
def equivalencia(i_t1, t2, t1):
    """Convierte tasa de período t1 a período t2"""
    return (1 + i_t1) ** (t2 / t1) - 1

# Ejemplo: tasa mensual a tasa anual
i_mensual = 0.03
i_anual = equivalencia(i_mensual, 12, 1)
print(f"Tasa mensual: {i_mensual*100}%")
print(f"Tasa anual equivalente: {i_anual*100:.2f}%")
```

### 3.2 Proporción (TNA ↔ Efectiva)

$TNA = i_{\text{efectiva}} \cdot \frac{365}{n}$

$i_{\text{efectiva}} = TNA \cdot \frac{n}{365}$

```python
def proporcion(TNA, n):
    """Convierte TNA a tasa efectiva del período n (en días)"""
    return TNA * (n / 365)

# Ejemplo: TNA 40%, equivalente a 90 días
TNA = 0.40
n = 90
efectiva_90d = proporcion(TNA, n)
print(f"TNA: {TNA*100}%")
print(f"Efectiva {n} días: {efectiva_90d*100:.4f}%")

# De efectiva a TNA
TNA_calculada = efectiva_90d / (n / 365)
print(f"TNA calculada: {TNA_calculada*100:.2f}%")
```

### 3.3 Valor Futuro (Capitalización Compuesta)

$C_n = C_0 \cdot (1 + i)^n$

```python
def capitalizacion_compuesta(C0, i, n):
    """Calcula valor futuro con capitalización compuesta"""
    return C0 * (1 + i) ** n

# Ejemplo: $1000 al 10% en 3 años
C0 = 1000
i = 0.10
n = 3
Cn = capitalizacion_compuesta(C0, i, n)
print(f"VF = ${C0} × (1 + {i})^{n} = ${Cn}")
```

### 3.4 Valor Presente

$C_0 = \frac{C_n}{(1 + i)^n}$

```python
def valor_presente(Cn, i, n):
    """Calcula valor presente (descuento)"""
    return Cn / ((1 + i) ** n)

# Ejemplo: ¿Cuánto vale hoy $1331 recibidos en 3 años al 10%?
Cn = 1331
C0 = valor_presente(Cn, 0.10, 3)
print(f"VP = ${Cn} / (1.10)^3 = ${C0}")
```

## 4. MÉTRICAS DE INVERSIÓN

### 4.1 Valor Actual Neto (VAN)

$VAN = \sum_{t=0}^{n} \frac{F_t}{(1+r)^t} - I_0$

- VAN > 0 → proyecto **rentable**
- VAN < 0 → **no rentable**
- VAN = 0 → indiferente

```python
import numpy_financial as nf

def calcular_van(flujos, tasa):
    """Calcula VAN usando numpy-financial
    flujos: lista donde flujos[0] = inversión inicial (negativo)
    tasa: tasa de descuento
    """
    return round(nf.npv(tasa, flujos), 2)

# Ejemplo 1
flujos_1 = [-10000, 3000, 3500, 4000, 4000, 2000]
tasa_1 = 0.50
van_1 = calcular_van(flujos_1, tasa_1)
print(f"VAN (ejemplo 1): ${van_1}")

# Ejemplo 2
flujos_2 = [-5000, 1500, 2000, 2000, 1000]
tasa_2 = 0.10
van_2 = calcular_van(flujos_2, tasa_2)
print(f"VAN (ejemplo 2): ${van_2}")
```

### 4.2 Tasa Interna de Retorno (TIR)

$\sum_{t=0}^{n} \frac{F_t}{(1+TIR)^t} = 0$

- TIR > r → rentable
- TIR < r → no rentable

```python
def calcular_tir(flujos):
    """Calcula TIR de una serie de flujos"""
    return round(nf.irr(flujos) * 100, 2)

# Ejemplo 1
flujos_e1 = [-10000, 3000, 4000, 5000, 3000]
tir_e1 = calcular_tir(flujos_e1)
print(f"TIR (ejemplo 1): {tir_e1}%")

# Ejemplo 2
flujos_e2 = [-20000, 5000, 4000, 3000, 3000, 2000, 2000, 1500, 1500, 1000, 1000]
tir_e2 = calcular_tir(flujos_e2)
print(f"TIR (ejemplo 2): {tir_e2}%")
```

### 4.3 Tasa Interna de Retorno Modificada (TIRM)

$TIRM = \left(\frac{VF_{\text{positivos}}}{VP_{\text{negativos}}}\right)^{1/n} - 1$

Asume reinversión a una tasa específica (no a la propia TIR como la TIR clásica).

```python
def calcular_tirm(flujos, tasa_financiamiento, tasa_reinversion):
    """Calcula TIRM
    tasa_financiamiento: tasa para descontar flujos negativos
    tasa_reinversion: tasa para capitalizar flujos positivos
    """
    return round(nf.mirr(flujos, tasa_financiamiento, tasa_reinversion) * 100, 2)

# Ejemplo
flujos_tirm = [-10000, 3000, 4000, 5000, 3000]
tirm = calcular_tirm(flujos_tirm, 0.10, 0.10)
print(f"TIR: {calcular_tir(flujos_tirm)}%")
print(f"TIRM (reinversión 10%): {tirm}%")
```

### 4.4 Índice de Rentabilidad (IR)

$IR = \frac{VAN}{\text{Inversión inicial}}$

- IR > 1 → rentable

```python
def calcular_ir(flujos, tasa):
    """Calcula Índice de Rentabilidad"""
    van = nf.npv(tasa, flujos)
    inversion = abs(flujos[0])
    return round(van / inversion, 2)

# Ejemplo
flujos_ir = [-10000, 3000, 4000, 5000, 3000]
ir = calcular_ir(flujos_ir, 0.10)
print(f"IR: {ir}")
print(f"¿Rentable? {'Sí' if ir > 1 else 'No'}")
```

### 4.5 Periodo de Recuperación Simple (PRS)

Suma de flujos brutos hasta igualar la inversión inicial. **No descuenta.**

```python
def calcular_prs(flujos):
    """Calcula Periodo de Recuperación Simple"""
    inversion = abs(flujos[0])
    acumulado = 0
    for t in range(1, len(flujos)):
        acumulado += flujos[t]
        if acumulado >= inversion:
            # Interpolación lineal
            excedente = acumulado - inversion
            fraccion = excedente / flujos[t]
            return round(t - 1 + (1 - fraccion), 2)
    return None  # No se recupera en el período

# Ejemplo
flujos_prs = [-10000, 3000, 4000, 5000, 3000]
prs = calcular_prs(flujos_prs)
print(f"PRS: {prs} años")
```

### 4.6 Periodo de Recuperación Descontado (PRD)

Igual que PRS pero con flujos descontados. **PRD siempre ≥ PRS.**

```python
def calcular_prd(flujos, tasa):
    """Calcula Periodo de Recuperación Descontado"""
    inversion = abs(flujos[0])
    acumulado = 0
    for t in range(len(flujos)):
        flujo_descontado = flujos[t] / (1 + tasa) ** t
        acumulado += flujo_descontado
        if acumulado >= inversion and t > 0:
            excedente = acumulado - flujo_descontado - (inversion - (acumulado - flujo_descontado))
            fraccion = (inversion - (acumulado - flujo_descontado)) / flujo_descontado
            return round((t - 1) + fraccion, 2)
    return None

# Ejemplo
flujos_prd = [-10000, 3000, 4000, 5000, 3000]
prd = calcular_prd(flujos_prd, 0.10)
print(f"PRD: {prd} años")
```

### 4.7 Fórmulas de Anualidades

$VP_{\text{anualidad}} = PMT \times \frac{1 - (1+r)^{-n}}{r}$

$VF_{\text{anualidad}} = PMT \times \frac{(1+r)^n - 1}{r}$

```python
def vp_anualidad(PMT, r, n):
    """Valor presente de una anualidad"""
    return PMT * (1 - (1 + r) ** (-n)) / r

def vf_anualidad(PMT, r, n):
    """Valor futuro de una anualidad"""
    return PMT * ((1 + r) ** n - 1) / r

# Ejemplo
PMT = 1000
r = 0.10
n = 5
print(f"VP anualidad: ${vp_anualidad(PMT, r, n):,.2f}")
print(f"VF anualidad: ${vf_anualidad(PMT, r, n):,.2f}")
```

## 5. COSTO DE CAPITAL (CAPM)

$K_o = r_f + \beta(r_m - r_f)$

Con riesgo país:

$K_o = r_f + \beta(r_m - r_f) + RP$

```python
def costo_capital(rf, beta, rm):
    """Calcula costo de capital Ke/Ko (CAPM)"""
    return rf + beta * (rm - rf)

def costo_capital_rp(rf, beta, rm, rp):
    """Calcula costo de capital con riesgo país"""
    return rf + beta * (rm - rf) + rp

# Ejemplo Integradora 2.2
rf = 0.04       # tasa libre de riesgo
beta = 1.45     # beta del proyecto
rm = 0.085      # retorno del mercado
rp = 0.1162     # riesgo país

ko = costo_capital(rf, beta, rm)
ko_rp = costo_capital_rp(rf, beta, rm, rp)

print(f"Ko anual: {ko*100:.3f}%")
print(f"Ko mensual: {((1 + ko)**(30/365) - 1)*100:.4f}%")
print(f"Ko con RP anual: {ko_rp*100:.3f}%")
print(f"Ko con RP mensual: {((1 + ko_rp)**(30/365) - 1)*100:.4f}%")
```

## 10. RESUMEN COMPARATIVO DE MÉTRICAS

| Métrica | Fórmula | Criterio | Descuenta flujos? | Función |
|---------|---------|----------|-------------------|----------|
| **VAN** | $\sum F_t/(1+r)^t - I_0$ | > 0 = rentable | Sí | `nf.npv()` |
| **TIR** | $\sum F_t/(1+TIR)^t = 0$ | > r = rentable | Sí | `nf.irr()` |
| **TIRM** | $(VF_+/VP_-)^{1/n} - 1$ | > r = rentable | Sí | `nf.mirr()` |
| **IR** | $VAN / I_0$ | > 1 = rentable | Sí | manual |
| **PRS** | Suma flujos brutos | Cuántos períodos | **No** | manual |
| **PRD** | Suma flujos descontados | Cuántos períodos | Sí | manual |
